```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic_df = pd.read_csv('https://raw.githubusercontent.com/IST407-707/707-lecture-master/e462aa3441a3948d3b64d6a5c423616bf4be6b02/2-week2/data/titanic.csv')

# Select numeric columns for scatter plot matrix
numeric_columns = titanic_df.select_dtypes(include=['int64', 'float64'])

# Create scatter plot matrix
sns.pairplot(numeric_columns)
plt.show()
```

#### **Exercise 2:**

It's zero cost to build a scatter matrix, but sometimes it only helps to highlight what you *don't* know. In addition to identifying correlations (or lack thereof) we'd also like to get a sense of the predictive power of different features.  Figure out how to use histogram plots to look at survival rates by age, pclass, and gender.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Assume 'df' is your DataFrame containing Titanic dataset or a similar dataset

# Set the style for Seaborn
sns.set(style="whitegrid")

# Plot histogram for Age by Survival
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x="age", hue="survived", multiple="stack", bins=30, kde=True)
plt.title('Survival Rates by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Plot histogram for Pclass by Survival
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic_df, x="pclass", hue="survived", multiple="stack", bins=3)
plt.title('Survival Rates by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# Plot histogram for Gender by Survival
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic_df, x="sex", hue="survived", multiple="stack", bins=2)
plt.title('Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
